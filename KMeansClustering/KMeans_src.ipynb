```python
K = 2
```

```python
clusters = []
for x in range (len(df)):
    clusters.append(np.random.randint(2))
print(clusters)
```

```python
df['Cluster'] = pd.Series(clusters)
df_scaled['Cluster'] = pd.Series(clusters)
df_scaled.head()
```

```python
df['Private'] = df['Private'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
def get_centroids(df):
    centroids = []
    for c in range(K):
        centroid = np.array(df[df['Cluster'] == c].iloc[:, :-1].mean())
        centroids.append(centroid)
    return centroids
```

```python
centroids = get_centroids(df.iloc[:, 2:])
centroids
```
